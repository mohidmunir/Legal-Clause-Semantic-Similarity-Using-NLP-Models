# Text Cleaning Utilities

```python
# ============================================================
#  Text Cleaning Utilities
# ============================================================
def clean_text(text: str) -> str:
    text = str(text)
    text = text.lower()
    text = re.sub(r'\d+', ' ', text)                       # remove digits
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def infer_text_column(df: pd.DataFrame) -> str:
    candidates = ["clause_text", "text", "clause", "Clause", "content"]
    for c in candidates:
        if c in df.columns: return c
    # fallback: choose object column with longest avg length
    objs = [c for c in df.columns if df[c].dtype == "object"]
    if not objs: raise ValueError("No textual column found.")
    return max(objs, key=lambda c: df[c].astype(str).str.len().mean())
```

# Load All Clauses Across CSVs

```python
# ============================================================
# ðŸ“¥ Load All Clauses Across CSVs (text, category)
# ============================================================
def load_all_clauses(data_dir: str, min_len: int = 6) -> pd.DataFrame:
    rows = []
    csv_files = sorted(glob.glob(os.path.join(data_dir, "*.csv")))
    print(f"ðŸ”Ž Found {len(csv_files)} CSV files.")
    skipped = 0
    for fp in csv_files:
        cat = os.path.splitext(os.path.basename(fp))[0]
        try:
            tmp = pd.read_csv(fp)
        except Exception as e:
            print(f"âš ï¸ Skipping {fp}: {e}")
            skipped += 1
            continue
        col = infer_text_column(tmp)
        tmp = tmp[[col]].dropna().rename(columns={col: "clause_text"})
        tmp["clean"] = tmp["clause_text"].map(clean_text)
        tmp = tmp[tmp["clean"].str.len() >= min_len]
        rows.extend([(t, cat) for t in tmp["clean"].tolist()])
    if not rows:
        raise RuntimeError("No clauses loaded. Check dataset/columns.")
    df = pd.DataFrame(rows, columns=["text", "category"])
    print(f"âœ… Loaded {len(df)} clauses from {len(csv_files)-skipped} files.")
    return df

all_df = load_all_clauses(DATA_DIR)
display(all_df.sample(5, random_state=SEED))
```

# Build Balanced Positive/Negative Pairs

```python
# ============================================================
#  Build Balanced Positive/Negative Pairs
#   Positive: same category, Negative: different categories
#   Balanced 1:1, limited per category for speed
# ============================================================
rng = np.random.default_rng(SEED)

def sample_pos_pairs(df, max_pairs_per_cat=3000):
    pos = []
    for cat, grp in df.groupby("category"):
        texts = grp["text"].to_list()
        if len(texts) < 2:
            continue
        n_pairs = min(max_pairs_per_cat, len(texts)//2)
        idx = rng.permutation(len(texts))
        for i in range(0, 2*n_pairs, 2):
            pos.append((texts[idx[i]], texts[idx[i+1]], 1))
    return pos

def sample_neg_pairs(df, n_needed):
    neg = []
    pools = {c: g["text"].to_list() for c, g in df.groupby("category")}
    cats = list(pools.keys())
    while len(neg) < n_needed:
        c1, c2 = rng.choice(cats, 2, replace=False)
        t1 = rng.choice(pools[c1]); t2 = rng.choice(pools[c2])
        neg.append((t1, t2, 0))
    return neg

pos_pairs = sample_pos_pairs(all_df, max_pairs_per_cat=2500)
neg_pairs = sample_neg_pairs(all_df, len(pos_pairs))

pairs_df = pd.DataFrame(pos_pairs + neg_pairs, columns=["clause1", "clause2", "label"])
pairs_df = pairs_df.sample(frac=1.0, random_state=SEED).reset_index(drop=True)

print(" Pairs built:", pairs_df.shape)
print("Class balance:", pairs_df["label"].value_counts().to_dict())
display(pairs_df.head())
```

# Tokenization & Padding (Shared for All Models)**bold text**

```python
# ============================================================
# Tokenization & Padding (Shared)
# ============================================================
VOCAB_SIZE = 20000
MAX_LEN = 120
OOV = "<OOV>"

tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token=OOV)
tokenizer.fit_on_texts(pairs_df["clause1"].tolist() + pairs_df["clause2"].tolist())

seq1 = tokenizer.texts_to_sequences(pairs_df["clause1"])
seq2 = tokenizer.texts_to_sequences(pairs_df["clause2"])

X1 = pad_sequences(seq1, maxlen=MAX_LEN, padding="post", truncating="post")
X2 = pad_sequences(seq2, maxlen=MAX_LEN, padding="post", truncating="post")
y  = pairs_df["label"].to_numpy(dtype=np.int32)

print("âœ… Tokenized:")
print("X1:", X1.shape, "X2:", X2.shape, "y:", y.shape)
```

# Stratified Train/Val Split

```python
# ============================================================
#  Stratified Train/Val Split
# ============================================================
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=SEED)
for tr, va in sss.split(X1, y):
    X1_tr, X1_va = X1[tr], X1[va]
    X2_tr, X2_va = X2[tr], X2[va]
    y_tr,  y_va  = y[tr],  y[va]

print("Train:", len(y_tr), " Val:", len(y_va))
print("Class balance (train):", np.bincount(y_tr))
print("Class balance (val):  ", np.bincount(y_va))
```

# Metrics & Plotting Helpers

```python
# ============================================================
#  Metrics & Plotting Helpers
# ============================================================
def compute_metrics(y_true, y_prob, threshold=0.5):
    y_pred = (y_prob >= threshold).astype(int)
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    try:
        roc = roc_auc_score(y_true, y_prob)
    except ValueError:
        roc = float("nan")
    ap = average_precision_score(y_true, y_prob)
    return {"Accuracy": acc, "Precision": prec, "Recall": rec, "F1": f1, "ROC-AUC": roc, "PR-AUC": ap}

def plot_training(history, title_prefix, save_name_prefix):
    # Accuracy if present
    if any("accuracy" in k for k in history.history.keys()):
        plt.figure()
        plt.plot(history.history.get("accuracy", []), label="train_acc")
        plt.plot(history.history.get("val_accuracy", []), label="val_acc")
        plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title(f"{title_prefix} â€“ Accuracy")
        plt.legend(); plt.grid(True, alpha=.3)
        plt.savefig(os.path.join(FIG_DIR, f"{save_name_prefix}_accuracy.png"), dpi=150)
        plt.show()

    # Loss
    plt.figure()
    plt.plot(history.history["loss"], label="train_loss")
    plt.plot(history.history["val_loss"], label="val_loss")
    plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title(f"{title_prefix} â€“ Loss")
    plt.legend(); plt.grid(True, alpha=.3)
    plt.savefig(os.path.join(FIG_DIR, f"{save_name_prefix}_loss.png"), dpi=150)
    plt.show()

def plot_confusion_matrix(y_true, y_prob, title, save_name, thr=0.5):
    y_pred = (y_prob >= thr).astype(int)
    cm = confusion_matrix(y_true, y_pred)
    plt.figure()
    plt.imshow(cm, cmap="Blues")
    plt.title(title); plt.colorbar()
    plt.xticks([0,1], ["Not Similar","Similar"]); plt.yticks([0,1], ["Not Similar","Similar"])
    for i in range(2):
        for j in range(2):
            plt.text(j, i, cm[i, j], ha="center", va="center")
    plt.xlabel("Predicted"); plt.ylabel("True")
    plt.tight_layout()
    plt.savefig(os.path.join(FIG_DIR, f"{save_name}_confusion_matrix.png"), dpi=150)
    plt.show()

def plot_pr_roc(y_true, y_prob, title_prefix, save_name_prefix):
    # PR
    p, r, _ = precision_recall_curve(y_true, y_prob)
    ap = average_precision_score(y_true, y_prob)
    plt.figure()
    plt.plot(r, p)
    plt.xlabel("Recall"); plt.ylabel("Precision")
    plt.title(f"{title_prefix} â€“ PR Curve (AP={ap:.3f})")
    plt.grid(True, alpha=.3)
    plt.savefig(os.path.join(FIG_DIR, f"{save_name_prefix}_pr_curve.png"), dpi=150)
    plt.show()

    # ROC
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    roc = roc_auc_score(y_true, y_prob)
    plt.figure()
    plt.plot(fpr, tpr)
    plt.xlabel("FPR"); plt.ylabel("TPR")
    plt.title(f"{title_prefix} â€“ ROC Curve (AUC={roc:.3f})")
    plt.grid(True, alpha=.3)
    plt.savefig(os.path.join(FIG_DIR, f"{save_name_prefix}_roc_curve.png"), dpi=150)
    plt.show()
```

# Model A â€“ Siamese BiLSTM (Distance + Dense)

```python
# ============================================================
#  Fixed Model A â€“ Siamese BiLSTM (Keras 3 compatible)
# ============================================================
from tensorflow.keras import layers, Model, Input

EMB_DIM = 128
LSTM_UNITS = 128
DROPOUT = 0.2

def build_siamese_bilstm(vocab_size, max_len, emb_dim=EMB_DIM, lstm_units=LSTM_UNITS, dropout=DROPOUT):
    inp1 = Input(shape=(max_len,), name="clause1")
    inp2 = Input(shape=(max_len,), name="clause2")

    emb = layers.Embedding(vocab_size, emb_dim, mask_zero=True, name="embedding")
    shared_bi = layers.Bidirectional(
        layers.LSTM(lstm_units, return_sequences=False, dropout=dropout),
        name="shared_bilstm"
    )

    x1 = shared_bi(emb(inp1))
    x2 = shared_bi(emb(inp2))

    # --- Use Keras layers for elementwise ops (no tf.math.* here) ---
    l1 = layers.Lambda(lambda t: tf.abs(t[0] - t[1]))([x1, x2])
    mult = layers.Multiply()([x1, x2])
    feat = layers.Concatenate()([l1, mult])

    x = layers.Dense(128, activation="relu")(feat)
    x = layers.Dropout(0.3)(x)
    out = layers.Dense(1, activation="sigmoid")(x)

    model = Model(inputs=[inp1, inp2], outputs=out, name="siamese_bilstm")
    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return model

#  Build successfully under TF 2.16 / Keras 3+
model_a = build_siamese_bilstm(VOCAB_SIZE, MAX_LEN)
model_a.summary()
```

# Training utilities

```python
# ============================================================
# ðŸƒ Training utilities (callbacks + one-line trainer)
# ============================================================
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

def common_callbacks(name_prefix):
    ckpt_path = os.path.join(ARTIFACT_DIR, f"{name_prefix}.keras")
    return [
        EarlyStopping(patience=3, restore_best_weights=True, monitor="val_loss"),
        ReduceLROnPlateau(patience=2, factor=0.5, monitor="val_loss", verbose=1),
        ModelCheckpoint(ckpt_path, save_best_only=True, monitor="val_loss", verbose=1),
    ]

def train_model(model, name_prefix, epochs=10, batch_size=256):
    history = model.fit(
        [X1_tr, X2_tr], y_tr,
        validation_data=([X1_va, X2_va], y_va),
        epochs=epochs, batch_size=batch_size, verbose=1,
        callbacks=common_callbacks(name_prefix)
    )
    plot_training(history, title_prefix=name_prefix, save_name_prefix=name_prefix)
    return model
```

# Train & evaluate Model A (Siamese BiLSTM)

```python
# ============================================================
#  Train & evaluate Model A (Siamese BiLSTM)
# ============================================================
model_a = build_siamese_bilstm(VOCAB_SIZE, MAX_LEN)
model_a = train_model(model_a, name_prefix="model_a_(siamese_bilstm)", epochs=10, batch_size=256)

# Validation predictions + metrics
y_prob_a = model_a.predict([X1_va, X2_va], batch_size=512).ravel()
metrics_a = compute_metrics(y_va, y_prob_a, threshold=0.5)
print(" Model A metrics:", metrics_a)

# Plots
plot_confusion_matrix(y_va, y_prob_a, "Model A â€“ Confusion Matrix", "model_a", thr=0.5)
plot_pr_roc(y_va, y_prob_a, "Model A", "model_a")
```

# Build Model B (BiLSTM + Additive Attention) â€” Keras 3â€“safe

```python
# ============================================================
#  Build Model B (BiLSTM + Additive Attention) â€“ Keras 3 safe
# ============================================================
class AdditiveAttention(layers.Layer):
    def __init__(self, units=128):
        super().__init__()
        self.W = layers.Dense(units, activation="tanh", use_bias=True)
        self.v = layers.Dense(1, use_bias=False)

    def call(self, seq, mask=None):
        # seq: (B, T, H)
        score = self.v(self.W(seq))  # (B, T, 1)
        if mask is not None:  # mask: (B, T)
            mask = tf.cast(mask, tf.float32)
            mask = tf.expand_dims(mask, axis=-1)  # (B, T, 1)
            score = score + (1.0 - mask) * (-1e9)
        weights = tf.nn.softmax(score, axis=1)   # (B, T, 1)
        context = tf.reduce_sum(weights * seq, axis=1)  # (B, H)
        return context

def build_bilstm_attention(vocab_size, max_len, emb_dim=128, lstm_units=128, dropout=0.2):
    # shared encoder with mask propagation
    tok_inp = Input(shape=(max_len,), name="tokens")
    emb = layers.Embedding(vocab_size, emb_dim, mask_zero=True, name="embedding")(tok_inp)
    seq = layers.Bidirectional(layers.LSTM(lstm_units, return_sequences=True, dropout=dropout))(emb)
    # mask flows automatically; Keras will pass it to the next layer
    ctx = AdditiveAttention(units=128)(seq)
    enc = layers.Dense(128, activation="relu")(ctx)
    enc = layers.Dropout(0.3)(enc)
    encoder = Model(tok_inp, enc, name="encoder_attention")

    c1 = Input(shape=(max_len,), name="clause1")
    c2 = Input(shape=(max_len,), name="clause2")
    e1 = encoder(c1); e2 = encoder(c2)
    feat = layers.Concatenate()([layers.Lambda(lambda t: tf.abs(t[0]-t[1]))([e1, e2]),
                                 layers.Multiply()([e1, e2])])
    h = layers.Dense(128, activation="relu")(feat)
    h = layers.Dropout(0.3)(h)
    out = layers.Dense(1, activation="sigmoid")(h)
    model = Model([c1, c2], out, name="bilstm_attention_pair")
    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return model

model_b = build_bilstm_attention(VOCAB_SIZE, MAX_LEN)
model_b.summary()
```

# Train & evaluate Model B

```python
# ============================================================
#  Train & evaluate Model B (BiLSTM + Attention)
# ============================================================
model_b = train_model(model_b, name_prefix="model_b_(bilstm_+_attention)", epochs=10, batch_size=256)

# Validation predictions + metrics
y_prob_b = model_b.predict([X1_va, X2_va], batch_size=512).ravel()
metrics_b = compute_metrics(y_va, y_prob_b, threshold=0.5)
print(" Model B metrics:", metrics_b)

# Plots
plot_confusion_matrix(y_va, y_prob_b, "Model B â€“ Confusion Matrix", "model_b", thr=0.5)
plot_pr_roc(y_va, y_prob_b, "Model B", "model_b")
```

```python
# ============================================================
# ðŸ”Ž Qualitative examples from the validation split
# ============================================================
# Recreate the same indices (same SEED -> identical split)
sss_tmp = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=SEED)
_, val_idx = next(sss_tmp.split(X1, y))
pairs_val = pairs_df.iloc[val_idx].reset_index(drop=True)

def sample_examples(y_true, y_prob, k=5, thr=0.5):
    y_pred = (y_prob >= thr).astype(int)
    idx = np.arange(len(y_true))
    correct = idx[y_pred == y_true]
    wrong   = idx[y_pred != y_true]
    rng.shuffle(correct); rng.shuffle(wrong)
    return correct[:k], wrong[:k]

def show_examples(name, pairs_subset, y_true, y_prob, k=5, thr=0.5):
    corr_idx, wrong_idx = sample_examples(y_true, y_prob, k=k, thr=thr)
    print(f"\n===== {name}: CORRECT ({len(corr_idx)}) =====")
    for i in corr_idx:
        print(f"[GT={y_true[i]}  P={y_prob[i]:.3f}]")
        print("A:", pairs_subset.iloc[i].clause1[:250])
        print("B:", pairs_subset.iloc[i].clause2[:250])
        print("-"*80)
    print(f"\n===== {name}: INCORRECT ({len(wrong_idx)}) =====")
    for i in wrong_idx:
        print(f"[GT={y_true[i]}  P={y_prob[i]:.3f}]")
        print("A:", pairs_subset.iloc[i].clause1[:250])
        print("B:", pairs_subset.iloc[i].clause2[:250])
        print("-"*80)

show_examples("Model A", pairs_val, y_va, y_prob_a, k=5, thr=0.5)
show_examples("Model B", pairs_val, y_va, y_prob_b, k=5, thr=0.5)
```
